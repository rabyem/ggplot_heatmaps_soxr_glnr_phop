```python
import pandas as pd
df = pd.read_csv("GlnR Transcriptomics Data TOPCAPI.csv", index_col= 'Locus_tag')
df
```

```python
df1 = df.stack().rename_axis(('a', 'b')).reset_index(name='val')
df1.rename(columns={"a": "Locus tag", "b": "Time", "val": "value"})
df2 = df1.rename(columns={"a": "Locus tag", "b": "Time", "val": "value"})
df2
```

```python
from plotnine import *
```

```python
colors = ("black", "white", "red")
lims= (0,50)
b = (5,25,50)
a = ggplot(df2, aes(x='Time', y='Locus tag')) + geom_tile(aes(fill = 'value'))+ scale_fill_gradientn(limits= lims,colors= colors, breaks=(5,25,50),labels=("5", "25", "50"))
print(a + ggtitle('GLNR REGULON M1146 TOPCAPI') + labs(y="Locus tag", x = "Time"))
```

```python
df3 = pd.read_csv("GlnR Transcriptomics Data TOPCAPI.csv", index_col= 'Locus_tag')
df4 = df3.transpose()
corr = df4.corr(method='pearson')
df5 = corr
df5
```

```python
df6 = df5.stack().rename_axis(('a', 'b')).reset_index(name='val')
df7 = df6.rename(columns={"a": "x", "b": "y", "val": "value"})
df7
```

```python
colors = ("blue", "cornflowerblue", "white", "pink", "red")
lims=(-1,1)
b = (-1, -0.5, 0, 0.5, 1)
a = ggplot(df7, aes(x='x', y='y')) + geom_tile(aes(fill = 'value'))+ scale_fill_gradientn(limits= lims,colors= colors, breaks=(-1,-0.5,0,0.5,1),labels=("-1", "-0.5", "0", "0.5", "1"))
```

```python
a + theme(axis_text_x = element_text(angle = 90, hjust=1))
```

```python
b = a + theme(axis_text_x = element_text(angle = 90, hjust=1))
b + ggtitle('GLNR REGULON M1146 TOPCAPI') + labs(y="Locus tag", x = "Locus tag")
```
